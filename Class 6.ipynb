```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15, 8))
plt.plot(df['id'], df['txn_total'], 'mo:')
plt.xlabel("Transaction ID")
plt.ylabel("Transaction amount in $")
plt.title("Line graph of txn_total")
plt.legend()

for x, y in zip(df['id'], df['txn_total']):
    label = "{:.2f}".format(y)
    plt.annotate(label,
               (x, y),
               textcoords = "offset points",
               xytext = (0, 10), 
               ha='center')
```

```python
# grouped bar graph/multiple-category bar graph:
# syntax
# plt.bar(xaxis_categories, bar_heights, width_per_hour)

width_per_bar = 0.4

course_list = ["CSE", "Mech", "ECE", "Aero", "Biomed"]

bar1 = np.arange(len(course_list))
bar2 = [(i+width_per_bar) for i in bar1]

boyscount_barheights = [150, 225, 60, 260, 300]
girlscount_barheights = [200, 100, 30, 120, 125]

plt.bar(bar1, boyscount_barheights, width_per_bar, label = "Boys")
plt.bar(bar2, girlscount_barheights, width_per_bar, label = "Girls")

plt.xticks(bar1 + width_per_bar/2, course_list)

plt.xlabel("Courses")
plt.ylabel("Number of students enrolled")
plt.title("Students vs Courses")

plt.legend()
```
